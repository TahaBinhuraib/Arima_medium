```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import chart_studio.plotly as py
import plotly.express as px
from datetime import datetime
import mplfinance as mpf
# !pip install statsmodels
import statsmodels.api as sm
# !pip install seaborn
import seaborn as sns
import plotly.graph_objects as go
# !pip install cufflinks
import cufflinks as cf 
from statsmodels.tsa.stattools import grangercausalitytests
```

```python
df_1 = pd.read_csv('US-Daily-Calories.csv')
```

```python
df_2 = pd.read_csv('US-Obese.csv')
df = pd.concat([df_1,df_2.obese] , axis = 1)
df.head()
```

```python
df['caloriesDiff'] = df['calories'].diff()
df['obeseDiff'] = df['obese'].diff()
df.dropna(inplace = True)
```

```python
df.head()
```

```python
#low p score means there is causality!
#f-test is aimed for smaller samples so better.
#chi2 test is more suitable for larger samples
grangercausalitytests(df[['obeseDiff', 'caloriesDiff']], maxlag=4)
```
